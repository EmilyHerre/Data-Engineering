Exercise 3 - Queries
1. Create a query that returns the top 3 car brands most sold (i.e., having the most car brand sales) during first and third quarter of year 2015.

```python
import pandas as pd

path_file = "CarSalesDataForReports (1).xlsx"
df_Clients = pd.read_excel(path_file, 'Clients')
df_Colors = pd.read_excel(path_file, 'Colors')
df_Stock = pd.read_excel(path_file, 'Stock')
df_Invoices = pd.read_excel(path_file, 'Invoices')
df_InvoiceLines = pd.read_excel(path_file, 'InvoiceLines')


#Merge the data from stock and invoiceLines into a dataframes

df_inventorySales = pd.merge(df_Stock, df_InvoiceLines, on=['StockID'])
df_inventorySales = pd.merge(df_inventorySales, df_Invoices, on=['InvoiceID'])

##Getting the invoices for quarter of year 2015
df_first_quarter_sales = df_inventorySales.loc[(df_inventorySales.InvoiceDateKey >= 20150101) & (df_inventorySales.InvoiceDateKey <= 20150331)] 
df_third_quarter_sales = df_inventorySales.loc[(df_inventorySales.InvoiceDateKey >= 20150701) & (df_inventorySales.InvoiceDateKey <= 20150930)] 


top_first_quarter = df_first_quarter_sales['Make'].value_counts().head(3).rename_axis('Brand').reset_index(name='Quantity')
top_third_quarter = df_third_quarter_sales['Make'].value_counts().head(3).rename_axis('Brand').reset_index(name='Quantity')
#Showing the first quarter and third quarter
print('**** FIRST QUARTER OF YEAR 2015 ****')
print(f'\n{top_first_quarter}\n')

print('**** THIRD QUARTER OF YEAR 2015 ****')
print(f'\n{top_third_quarter}\n')
```

2. Create a query that shows the top 3 most sold car colors (i.e., having the most car color sales) for each quarter for the years 2012,2013,2014,2015. 

```python
def show_quarter(data, year, quarter):
    print(f'\n ***{quarter} QUARTER OF {year} *** \n')
    print(f'{data}\n')


## method for find the color most sold by quarters
def findTopSellingCarColorsByQuarter(years):
    df_inventorySales = pd.merge(df_Stock, df_InvoiceLines, on=['StockID'])
    df_inventorySales = pd.merge(df_inventorySales, df_Invoices, on=['InvoiceID'])
    df_inventorySales = pd.merge(df_inventorySales, df_Colors, on=['ColorID'])

    ##loop over years
    for year in years:

        df_first_quarter_sales = df_inventorySales.loc[(df_inventorySales.InvoiceDate >= year + '-01-01') & (df_inventorySales.InvoiceDate <= year + '-03-31')] 
        top_first_quarter = df_first_quarter_sales['Color'].value_counts().head(3).reset_index(name='Quantity')
        show_quarter(top_first_quarter, year, 'FIRST')

        df_second_quarter_sales = df_inventorySales.loc[(df_inventorySales.InvoiceDate >= year + '-04-01') & (df_inventorySales.InvoiceDate <= year + '-06-30')] 
        top_second_quarter = df_second_quarter_sales['Color'].value_counts().head(3).reset_index(name='Quantity')
        show_quarter(top_second_quarter, year, 'SECOND')

        df_third_quarter_sales = df_inventorySales.loc[(df_inventorySales.InvoiceDate >= year + '-07-01') & (df_inventorySales.InvoiceDate <= year + '-09-30')] 
        top_third_quarter = df_third_quarter_sales['Color'].value_counts().head(3).reset_index(name='Quantity')
        show_quarter(top_third_quarter, year, 'THIRD')

        df_fourth_quarter_sales = df_inventorySales.loc[(df_inventorySales.InvoiceDate >= year + '-10-01') & (df_inventorySales.InvoiceDate <= year + '-12-31')] 
        top_fourth_quarter = df_fourth_quarter_sales['Color'].value_counts().head(3).reset_index(name='Quantity')
        show_quarter(top_fourth_quarter, year, 'FOURTH')


findTopSellingCarColorsByQuarter(['2012', '2013', '2014', '2015'])
```
